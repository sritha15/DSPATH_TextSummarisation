```python
import pandas as pd
from rouge import Rouge

def calculate_rouge(hypothesis, reference):
    rouge = Rouge()
    scores = rouge.get_scores(hypothesis, reference, avg=True)

    formatted_scores = {
        'llmTR1-R': scores['rouge-1']['r'],
        'llmTR1-P': scores['rouge-1']['p'],
        'llmTR1-F': scores['rouge-1']['f'],
        'llmTR2-R': scores['rouge-2']['r'],
        'llmTR2-P': scores['rouge-2']['p'],
        'llmTR2-F': scores['rouge-2']['f'],
        'llmTRL-R': scores['rouge-l']['r'],
        'llmTRL-P': scores['rouge-l']['p'],
        'llmTRL-F': scores['rouge-l']['f']
    }
    return formatted_scores

results_df = pd.read_csv("/content/finetunedllm_summaries.csv")

rouge_scores = results_df.apply(lambda row: calculate_rouge(row['Finetuned LLM Summary'], row['Highlight']), axis=1, result_type='expand')

results_df = pd.concat([results_df, rouge_scores], axis=1)

pd.set_option('display.max_columns', None)

print(results_df)
```

```python
results_df.to_csv('finetunedllmwithrouges.csv', index=False)
```

```python
import pandas as pd
from rouge import Rouge

def calculate_rouge(hypothesis, reference):
    rouge = Rouge()
    scores = rouge.get_scores(hypothesis, reference, avg=True)

    formatted_scores = {
        'llmTC_R1-R': scores['rouge-1']['r'],
        'llmTC_R1-P': scores['rouge-1']['p'],
        'llmTC_R1-F': scores['rouge-1']['f'],
        'llmTC_R2-R': scores['rouge-2']['r'],
        'llmTC_R2-P': scores['rouge-2']['p'],
        'llmTC_R2-F': scores['rouge-2']['f'],
        'llmTC_RL-R': scores['rouge-l']['r'],
        'llmTC_RL-P': scores['rouge-l']['p'],
        'llmTC_RL-F': scores['rouge-l']['f']
    }
    return formatted_scores

results_df = con

rouge_scores = results_df.apply(lambda row: calculate_rouge(row['Finetuned LLM Summary With Context'], row['Highlight']), axis=1, result_type='expand')

results_df = pd.concat([results_df, rouge_scores], axis=1)

pd.set_option('display.max_columns', None)

results_df.head()
```
